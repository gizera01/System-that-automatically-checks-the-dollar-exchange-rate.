```python
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

#remove automation warnings from the screen
chrome_options = Options()
chrome_options.add_argument("--disable-infobars")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])

navegador = webdriver.Chrome(options=chrome_options)

try:
    #Access the page AwesomeAPI
    url_publica = "https://economia.awesomeapi.com.br/json/last/USD-BRL"
    navegador.get(url_publica)
    
    #2-second pause
    time.sleep(2)
    
    #captures all displayed text
    texto_da_tela = navegador.find_element(By.TAG_NAME, "body").text
    
    #converts JSON to a dictionary
    dados_cotacao = json.loads(texto_da_tela)
    
    #Get the dollar-to-real exchange rate
    valor_dolar = dados_cotacao["USDBRL"]["bid"]
    
    #displays the result in the terminal
    print(f"Valor do dólar hoje: R$ {float(valor_dolar):.2f}")

except Exception as e:
    print(f"ocorreu um erro: {e}")

finally:
    #Close the browser.
    navegador.quit()
```
